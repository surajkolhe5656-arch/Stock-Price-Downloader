```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

data = yf.download("INFY.NS", start="2024-01-01", end="2025-01-01")

print(data.head())
```

```python
# 20-day and 50-day moving averages
data["MA20"] = data["Close"].rolling(20).mean()
data["MA50"] = data["Close"].rolling(50).mean()
```

```python
data["Daily_Return"] = data["Close"].pct_change()
```

```python
plt.figure(figsize=(12,6))
plt.plot(data.index, data["Close"], label="INFY Close Price", color="blue")
plt.plot(data.index, data["MA20"], label="20-day MA", color="orange")
plt.plot(data.index, data["MA50"], label="50-day MA", color="green")
plt.title("Infosys Stock Price with Moving Averages")
plt.xlabel("Date")
plt.ylabel("Price (INR)")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(data.index, data["Daily_Return"], label="Daily Returns", color="red")
plt.title("Infosys Daily Returns")
plt.xlabel("Date")
plt.ylabel("Return")
plt.legend()
plt.show()
```
